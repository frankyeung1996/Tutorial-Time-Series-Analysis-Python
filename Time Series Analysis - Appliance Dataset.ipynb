```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#Loading statsmodel method to perform decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Reading data
#Appliance = pd.read_csv('Appliance.csv')

# Another way to load a time series data by defining a date format
dateparse = lambda dates: pd.datetime.strptime(dates, '%m/%d/%Y %H:%M')
Appliance = pd.read_csv('Appliance.csv', index_col='date', date_parser=dateparse)

Appliance.resample('M').mean().head()

Appliance.head()
```

```python
# Visualizing trend of Appliance data
plt.plot(seasonal_decompose(Appliance,freq=5000).trend)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Number of appliances', fontsize=15)
plt.title('Number of appliances present against the date', weight='bold', fontsize=18)
plt.show()
```

```python
# Visualizing seasonality of Appliance data
plt.plot(seasonal_decompose(Appliance,freq=1000).seasonal)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Number of appliances', fontsize=15)
plt.title('Number of appliances present against the date', weight='bold', fontsize=18)
plt.show()
```

```python
plt.plot(seasonal_decompose(Appliance,freq=10).resid)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Number of appliances', fontsize=15)
plt.title('Number of appliances present against the date', weight='bold', fontsize=18)
plt.show()
```
